### Описание исходных данных

***Интернет-магазин «В один клик» продаёт разные товары: для детей, для дома, мелкую бытовую технику, косметику и даже продукты.
```python
def col_display(text, color="black", bold=False, bgcolor=None):
    # Добавляем пробел в начале текста
    text = "'" + text

    # Стиль для текста
    style = f"color: {color};"
    
    # Добавление жирности, если нужно
    if bold:
        style += " font-weight: bold;"
    
    # Добавление фона, если указан
    if bgcolor:
        style += f" background-color: {bgcolor};"
    
    # Устанавливаем margin и padding, чтобы избежать обрезки текста
    style += " padding: 0; margin: 0; display: inline;"

    # Формируем HTML для вывода
    html_text = f"<div style='{style}'>{text}</div>"
    
    # Отображаем текст
    display(HTML(html_text))
    
def err_display(text):
    col_display('ERROR: ' + text, color='white', bgcolor="red", bold=True)
    
def info_display(text):
    col_display('INFO: ' +  text, color='white', bgcolor="blue", bold=True)
    
def cycle_display(text):
    col_display('Текущая итерация: ' +  text, color='yellow', bold=True)
    
def final_display(text):
    col_display(text, color='green', bold=True)    
    
def title_display(text):
    col_display(text, color='white', bgcolor="green", bold=True)
```

```python
# Пути к файлам
market_file_path = 'https://code.s3.yandex.net/datasets/market_file.csv'
market_money_path = 'https://code.s3.yandex.net/datasets/market_money.csv'
market_time_path = 'https://code.s3.yandex.net/datasets/market_time.csv'
money_path = 'https://code.s3.yandex.net/datasets/money.csv'

# Функция для безопасной загрузки данных
def load_data(file_path, sep=',', decimal='.'):
    try:
        data = pd.read_csv(file_path, sep=sep, decimal=decimal)
        info_display(f"Успешно загружен файл: {file_path}")
        return data
    except FileNotFoundError:
        err_display(f"Ошибка: Файл не найден - {file_path}")
    except pd.errors.ParserError:
        err_display(f"Ошибка: Неправильный формат данных в файле - {file_path}")
    except Exception as e:
        err_display(f"Неизвестная ошибка при загрузке {file_path}: {e}")

# Загрузка данных
market_file = load_data(market_file_path)  # По умолчанию запятая
market_money = load_data(market_money_path)  # По умолчанию запятая
market_time = load_data(market_time_path)  # По умолчанию запятая
money = load_data(money_path, sep=';', decimal=',')  # Для money.csv используем точку с запятой

# Проверка загруженных данных, если они успешны
datasets = {
    "market_file": market_file,
    "market_money": market_money,
    "market_time": market_time,
    "money": money
}

for name, data in datasets.items():
    if data is not None:
        cycle_display(f"Данные из {name}:")
        display(data.head())
    else:
        err_display(f"{name} не был загружен из-за ошибок.")
```

***Переименовали столбцы для более удобной работы с ними. Также вывели количество пропусков и явных дубликатов для каждого датасета***

```python
def rename_columns_with_regex(df):
    # Преобразование названий столбцов
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace(r'\s+', '_', regex=True)
    df.columns = df.columns.str.replace(r'\W+', '', regex=True)
    return df

def handle_missing_data(df, dataset_name):
    # Выводим количество пропусков в каждом столбце
    missing_data = df.isnull().sum()
    if missing_data.sum() == 0:
        final_display(f"Пропуски в датасете {dataset_name} отсутствуют.")
    else:
        cycle_display(f"Количество пропусков в каждом столбце датасета {dataset_name}:")
        final_display(missing_data)
    return df

def preprocess_data(market_file, market_money, market_time, money):
    # Применяем регулярные выражения для переименования столбцов
    if market_file is not None:
        market_file = rename_columns_with_regex(market_file)

    if market_money is not None:
        market_money = rename_columns_with_regex(market_money)

    if market_time is not None:
        market_time = rename_columns_with_regex(market_time)

    if money is not None:
        money = rename_columns_with_regex(money)

    if market_file is not None:
        market_file = handle_missing_data(market_file, "market_file")

    if market_money is not None:
        market_money = handle_missing_data(market_money, "market_money")

    if market_time is not None:
        market_time = handle_missing_data(market_time, "market_time")

    if money is not None:
        money = handle_missing_data(money, "money")
        
    # Выводим информацию о данных после предобработки
    title_display("Проверка предобработанных данных:")
    datasets = {
        "market_file": market_file,
        "market_money": market_money,
        "market_time": market_time,
        "money": money
    }

    for name, data in datasets.items():
        if data is not None:
            cycle_display(f"Данные из {name}:")
            data.info()
        else:
            err_display(f"{name} не был загружен из-за ошибок.")
    
    # Вывод явных дубликатов
    title_display("Поиск явных дубликатов в данных:")
    for name, data in datasets.items():
        if data is not None:
            duplicates = data[data.duplicated()]
            if not duplicates.empty:
                cycle_display(f"Дубликаты в {name}:")
                display(duplicates)
            else:
                final_display(f"Явных дубликатов в {name} не найдено.")
    

# Вызов функции предобработки данных
preprocess_data(market_file, market_money, market_time, money)
```

```python
# Функция для изменения типов данных
def convert_data_types(df):
    # Преобразуем столбцы в категориальные
    categorical_columns = ['покупательская_активность', 'тип_сервиса', 'разрешить_сообщать', 'популярная_категория']
    for column in categorical_columns:
        if column in df.columns:
            df[column] = df[column].astype('category')

    # Преобразуем числовые столбцы
    numeric_columns = ['маркет_актив_6_мес', 'маркет_актив_тек_мес', 'длительность', 'акционные_покупки',
                       'средний_просмотр_категорий_за_визит', 'неоплаченные_продукты_штук_квартал', 
                       'ошибка_сервиса', 'страниц_за_визит']
    for column in numeric_columns:
        if column in df.columns:
            df[column] = pd.to_numeric(df[column], errors='coerce')


    return df

# Вызов функции для изменения типов данных
market_file = convert_data_types(market_file)

# Проверим типы данных после изменений
display(market_file.dtypes)
```

## Объединение таблиц

```python
def merge_tables_correctly(market_file, market_money, market_time, money):
    """
    Объединение таблиц с учетом различных периодов, чтобы каждая строка соответствовала одному id,
    а данные о выручке и времени распределились по отдельным столбцам для каждого периода.
    """
    # Создаем pivot для market_money (выручка) и market_time (время на сайте)
    revenue_pivot = market_money.pivot(index='id', columns='период', values='выручка').reset_index()
    time_pivot = market_time.pivot(index='id', columns='период', values='минут').reset_index()

    # Переименуем столбцы для удобства
    revenue_pivot.columns = ['id'] + [f"выручка_{col}" for col in revenue_pivot.columns[1:]]
    time_pivot.columns = ['id'] + [f"время_{col}" for col in time_pivot.columns[1:]]

    # Объединяем таблицы pivot с market_file и money
    final_data = market_file.merge(revenue_pivot, on='id', how='left') \
                            .merge(time_pivot, on='id', how='left') \
                            .merge(money, on='id', how='left')

    # Удаляем лишние столбцы, если они остались
    final_data = final_data.drop(columns=['период_x', 'период_y'], errors='ignore')

    return final_data


# Вызов функции объединения таблиц
merge_data = merge_tables_correctly(market_file, market_money, market_time, money)

# Вывод результата
final_display(f"Размер итоговой таблицы: {merge_data.shape}")
display(merge_data.head())
```

## Корреляционный анализ

```python
# Настройки цветов и границ
colors = ["lightblue", "yellow", "red"]  # Цветовая шкала
bounds = [0.3, 0.5, 0.7, 1.0]  # Границы значений для отображения
cmap = ListedColormap(colors)
norm = plt.Normalize(vmin=0.3, vmax=1.0)

# Функция для отображения кастомной тепловой карты
def plot_custom_heatmap(corr_matrix, title):
    """
    Отображает корреляционную матрицу с настраиваемыми цветами.
    """
    plt.figure(figsize=(10, 8))
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  # Маска для верхнего треугольника
    sns.heatmap(corr_matrix, annot=True, cmap=cmap, norm=norm, mask=mask, cbar=True, 
                cbar_kws={'boundaries': bounds, 'ticks': bounds}, fmt='.2f')
    plt.title(title)
    plt.show()

# Уникальные классы признака "покупательская_активность"
unique_classes = merge_data['покупательская_активность'].unique()

# Количественные столбцы
quantitative_cols = merge_data.select_dtypes(include=['float64', 'int64']).drop(columns='id').columns

# Рассчитаем корреляции для общего файла без кодирования
corr_phik_file = merge_data.phik_matrix(interval_cols=quantitative_cols)

# Построим тепловую карту для merge_file
plot_custom_heatmap(corr_phik_file, "Phik Correlation Matrix for All Data")

# Рассчитываем и визуализируем корреляции для каждого класса
for activity_level in unique_classes:
    class_data = merge_data[merge_data['покупательская_активность'] == activity_level]
    corr_phik_class = class_data.phik_matrix(interval_cols=quantitative_cols)
    
    # Визуализация для текущего класса
    plot_custom_heatmap(corr_phik_class, f"Phik Correlation Matrix for Activity Level {activity_level}")
```

```python
try:
    RANDOM_STATE = 42
    TEST_SIZE = 0.25

    # Разделяем целевую переменную и признаки
    X_train, X_test, y_train, y_test = train_test_split(
        merge_data.drop(['покупательская_активность', 'id', 'прибыль'], axis=1),
        merge_data['покупательская_активность'],
        test_size=TEST_SIZE,
        random_state=RANDOM_STATE,
        stratify=merge_data['покупательская_активность']
    )

    # Определяем признаки
    ohe_columns = ['популярная_категория']
    ord_columns = ['акционные_покупки', 'тип_сервиса', 'разрешить_сообщать']
    num_columns = merge_data.select_dtypes(include=['float64', 'int64']).drop(columns=['id', 'прибыль']).columns.tolist()

    info_display('Ячейка загрузилась корректно')
except:
    err_display('При загрузке ячейки произошла ошибка')
```

***Пайплайн для предобработки данных***

```python
try:
    # Пайплайн для OneHotEncoder
    ohe_pipe = Pipeline([
        ('imputer', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
    ])

    # Пайплайн для OrdinalEncoder
    ord_pipe = Pipeline([
        ('imputer', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)),
        ('fill_missing', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ])
    info_display('Ячейка загрузилась корректно')
except:
    err_display('При загрузке ячейки произошла ошибка')
```

```python
try:
    # Общий процессор данных
    data_preprocessor = ColumnTransformer([
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', MinMaxScaler()),
            ('imputer', SimpleImputer(strategy='median'))
        ]), num_columns)
    ], remainder='passthrough')

    # Итоговый пайплайн
    pipe_final = Pipeline([
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))  # Заглушка, заменяется при поиске
    ])
    info_display('Ячейка загрузилась корректно')
except:
    err_display('При загрузке ячейки произошла ошибка')
```

***Т.к магазину очень важно правильно предсказывать класс `снизилась`, для выбора модели будем использовать пользовательскую метрику - `количество ложноотрицальных прогнозов`*** 

```python
def false_negatives(y_true, y_pred):
    conf_matrix = confusion_matrix(y_true, y_pred)
    return conf_matrix[1][0] / (conf_matrix[1][0] + conf_matrix[1][1])
```

***Удалили малозначимые признаки для лучшей работы модели***

```python
# Листы для сохранения результатов
results = []

# Перебор различных значений k (количество признаков)
for k in range(1, X_train_preprocessed.shape[1] + 1):  # От 1 до количества всех признаков
    # Шаг 2: Выбор K лучших признаков
    k_best_selector = SelectKBest(score_func=mutual_info_classif, k=k)
    X_train_selected = k_best_selector.fit_transform(X_train_preprocessed, y_train)
    X_test_selected = k_best_selector.transform(X_test_preprocessed)
    
    # Шаг 3: Обучение модели
    true_model.fit(X_train_selected, y_train)
    
    # Прогноз вероятностей для тестовой выборки
    y_test_pred = true_model.predict_proba(X_test_selected)[:, 1]
    
    # Вычисление ROC-AUC
    roc_auc = roc_auc_score(y_test, y_test_pred)
    
    # Прогноз классов
    y_test_pred_classes = (y_test_pred >= 0.5).astype(int)
    
    # Ложноотрицательная метрика
    false_negatives = -fn_scorer(true_model, X_test_selected, y_test)
    
    # Сохранение результата
    results.append((k, roc_auc, false_negatives))
    cycle_display(f"k={k}, ROC-AUC={roc_auc:.4f}, Ложноотрицательная метрика={false_negatives}")

# Выбор лучшего результата
best_result = min(results, key=lambda x: x[2])  # Минимизируем ложноотрицательную метрику
best_k, best_roc_auc, best_false_negatives = best_result

title_display("\nЛучший результат:")
title_display(f"Количество признаков (k): {best_k}")
title_display(f"ROC-AUC: {best_roc_auc:.4f}")
title_display(f"Ложноотрицательная метрика: {best_false_negatives}")

# Получение отобранных признаков для лучшего k
k_best_selector = SelectKBest(score_func=mutual_info_classif, k=best_k)
k_best_selector.fit(X_train_preprocessed, y_train)
selected_features = [X_train_preprocessed.columns[i] for i in k_best_selector.get_support(indices=True)]
title_display(f"Лучшие признаки: {selected_features}")
```

## Сегментация клиентов

***Чтобы сегментировать покупателей на основе вероятности снижения спроса и прибыли, а также проанализировать различия в группах покупателей, можно использовать следующие шаги. Давайте рассмотрим процесс подробнее:***

### Сегментация покупателей
#### Получение данных о вероятности снижения и прибыльности покупателей:

- ***В качестве переменных для сегментации будем использовать:***
    * `Вероятность снижения (вероятность того, что клиент станет неактивным).`
    * `Прибыль (например, средний доход от покупок клиента за определенный период времени).`
    
#### Определение порогов для сегментации:

- ***Чтобы разделить покупателей на различные сегменты, определим два порога:***
    * ***`Порог вероятности снижения` — это значение, выше или ниже которого клиент считается вероятно потерянным или стабильным.***
    * ***`Порог рентабельности` — это значение, выше которого покупатель считается прибыльным, а ниже — неэффективным с точки зрения прибыли.***
- ***Например, если вероятность снижения выше 0,5, а прибыль ниже определённого порога (например, 1000 рублей), клиент может быть отнесён к группе риска, с которой нужно работать более активно.***

#### Визуализация для определения порогов:

- ***Построим график рассеяния (scatter plot), где ось X будет показывать прибыль, а ось Y  вероятность снижения. Пороги для этих переменных будут разделять пространство на 4 сегмента:***
    * ***`Сегмент 1 (высокая прибыль, низкая вероятность снижения):` это идеальные клиенты, с которыми стоит продолжать работать.***
    * ***`Сегмент 2 (Низкая прибыль, низкая вероятность снижения):` эти клиенты стабильны, но не приносят много прибыли.***
    * ***`Сегмент 3 (высокая прибыль, высокая вероятность снижения):` это клиенты, которые приносят много прибыли, но риск их потери высок.***
    * ***`Сегмент 4 (Низкая прибыль, высокая вероятность снижения):` эти клиенты рискованные и невыгодные, с ними нужно работать, чтобы удержать их, или рассматривать вариант исключения из базы.***

***Для анализа выбрали 3 сегмент, как наиболее перспективный. Под высокой прибылью подразумеваем все, что больше медианы по всему датасету(т.е их прибыль больше, чем у 50% датасета)***

```python
# Удаляем ненужные столбцы, такие как 'пользовательская_активность' и 'id'
columns_to_drop = ['пользовательская_активность', 'id']
processed_data = merge_data.drop(columns=columns_to_drop, errors='ignore')


# Удаляем малозначимые столбцы (если они известны) ДО трансформации
processed_data = processed_data.drop(columns=unselected_features, errors='ignore')

# Преобразуем данные с использованием preprocessor
processed_data_array = preprocessor.transform(processed_data)

# Преобразуем обратно в DataFrame для совместимости
processed_data = pd.DataFrame(processed_data_array, columns=preprocessor.get_feature_names_out())

# Добавляем предсказания модели
merge_data['вероятность_снижения'] = true_model.predict_proba(processed_data)[:, 1]

display(merge_data.head())
```

```python
# Устанавливаем пороги для сегментации
threshold_churn = 0.5  # Порог вероятности снижения
threshold_profit = merge_data['прибыль'].median()  # Порог прибыли

# Создаем новый сегмент: высокая прибыль и низкая вероятность снижения
merge_data['high_profit_low_churn'] = ((merge_data['вероятность_снижения'] > threshold_churn) & 
                                       (merge_data['прибыль'] > threshold_profit))

# Построение графика рассеяния
plt.figure(figsize=(12, 6))
sns.scatterplot(data=merge_data, 
                x='вероятность_снижения', 
                y='прибыль', 
                hue='high_profit_low_churn', 
                palette={True: 'blue', False: 'purple'}, 
                legend='full')
plt.axvline(x=threshold_churn, color='red', linestyle='--', label='Порог вероятности снижения')
plt.axhline(y=threshold_profit, color='red', linestyle='--', label='Порог прибыли')
plt.title("Сегментация покупателей по прибыли и вероятности снижения")
plt.xlabel("Вероятность снижения")
plt.ylabel("Прибыль")
plt.legend()
plt.show()
```

***Сравнили распределения для каждого признака для интересующего сегмента и остальных данных***

```python
# Определяем сегмент с высокой прибылью и низкой вероятностью снижения
selected_segment = merge_data[(merge_data['вероятность_снижения'] < 0.5) & (merge_data['прибыль'] > 4.0)]
non_selected_segment = merge_data[~((merge_data['вероятность_снижения'] < 0.5) & (merge_data['прибыль'] > 4.0))]


# Убираем признаки, которые не анализируем
columns_to_analyze = [col for col in merge_data.columns if col not in ['id', 'пользовательская_активность', 'вероятность_снижения', 'прибыль', 'high_profit_low_churn']]

# Добавляем колонку для hue, если ее нет
merge_data['high_profit_low_churn'] = (merge_data['вероятность_снижения'] < 0.5) & (merge_data['прибыль'] > 4.0)

# Цикл по всем признакам для построения распределений
for feature in columns_to_analyze:
    # Определяем тип признака
    if merge_data[feature].dtype in ['float64', 'int64']:  # Числовые признаки
        plt.figure(figsize=(10, 5))
        sns.histplot(selected_segment[feature], color='blue', label='Высокая прибыль и низкая активность', kde=True, bins=30, stat='density')
        sns.histplot(non_selected_segment[feature], color='purple', label='Остальные клиенты', kde=True, bins=30, stat='density')
        plt.title(f"Распределение: {feature}")
        plt.xlabel(f"Значение {feature}")
        plt.ylabel("Плотность")
        plt.legend()
        plt.tight_layout()  # Автоматическая настройка размещения элементов
        plt.show()
    elif merge_data[feature].dtype in ['object', 'category', 'CategoricalDtype']:  # Категориальные признаки
        plt.figure(figsize=(10, 5))
        ax = sns.countplot(data=merge_data, x=feature, hue='high_profit_low_churn', palette={True: 'blue', False: 'purple'})
        plt.title(f"Распределение частот: {feature}")
        plt.xlabel(f"Значение {feature}")
        plt.ylabel("Частота")
        plt.legend(title='Сегмент', labels=['Высокая прибыль и низкая активность', 'Остальные клиенты'])
        
        # Автоматически вращаем подписи на оси X, чтобы избежать наложения
        plt.xticks(rotation=45, ha="right")  # Поворот подписей на 45 градусов, выравнивание справа
        plt.tight_layout()  # Автоматическая настройка размещения элементов
        plt.show()
    else:
        err_display(f"Пропускаем неизвестный тип признака: {feature}")
```
